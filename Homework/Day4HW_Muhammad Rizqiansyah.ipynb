```python
from typeguard import typechecked
@typechecked

def is_leap(year): # Menentukan tahun kabisat berdasarkan kalender Gregorian
    return (year % 400 == 0) | (year % 4 == 0 and year % 100 != 0)

def leap_year(x: int):
    if is_leap(x): # Menampilkan jika x adalah tahun kabisat
        print('Year',x,'is a leap year')
    y = x
    while not is_leap(y): # Mencari tahun kabisat berikutnya jika x bukan tahun kabisat
        y += 1
        if y % 4 == 0: # Menampilkan jika y adalah tahun kabisat
            print('Year',x,'is not a leap year. The next leap year after',x,'is',y)
```

```python
leap_year(2001) 
```

```python
leap_year(2004)
```
