```python
# No. 1 Membuat DataFrame dari Dictionary
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "name": ["Anna", "Dane", "David", "Kevin", "Joe", "Rose"],
    "weight": [45.2, 50.5, 62.1, 64.0, 59.0, np.nan],
    "height": [155, 160, 162, 170, 167, 159],
    "birth_year": [2000, 1997, 1996, 1987, 1998, 2003],
    "gender": ["female", "male", "male", "male", "male", "female"],
    "city": ["New York", "Chicago", "Austin", "New York", np.nan, "New York"]
})

df
```

```python
# No. 2 Menampilkan info DataFrame dasar
df.info()
```

```python
# No. 3 Menampilkan informasi statistik dasar untuk kolom numerik
df.describe()
```

```python
# No. 4 Menyusun ulang urutan kolom
df = df[["name", "birth_year", "height", "weight", "city", "gender"]]

df
```

```python
# No. 5 Menampilkan filter DataFrame dengan dua kondisi ((AND)OR(AND))
df[((df["gender"] == "male") & (df["birth_year"] > 1996)) | ((df["gender"] == "female") & (df["height"] > 157))]
```

```python
# No 6 Menangani Missing Values

# Mengisi NaN dari berat badan dengan rata-ratanya dari tiap gender
df['weight'] = df['weight'].fillna(df.groupby('gender')['weight'].transform('mean'))
# Mengisi NaN dari kota dengan modusnya
df["city"] = df["city"].fillna(df["city"].mode()[0])

df
```

```python
# No 7 Mengecek outliers dari berat badan (IQR)
q1 = df["weight"].quantile(0.25)
q3 = df["weight"].quantile(0.75)

iqr = q3-q1 #Interquartile range
fence_low  = q1-1.5*iqr
fence_high = q3+1.5*iqr

df.loc[(df["weight"] < fence_low) | (df["weight"] > fence_high)]
```

```python
# No 7 Mengecek outliers dari berat badan (Z-Score)
import scipy.stats as stats

df[(np.abs(stats.zscore(df["weight"])) >= 3)]
```
