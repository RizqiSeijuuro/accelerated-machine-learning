```python
# No 1 Membuat Pandas DataFrame dari Dictionary
import pandas as pd
import numpy as np

df1 = pd.DataFrame({
    "name": ["Anna", "Dane", "David", "Kevin", "Joe", "Rose"],
    "weight": [45.2, 50.5, 62.1, 64.0, 59.0, np.nan],
    "height": [155, 160, 162, 170, 167, 159],
    "birth_year": [2000, 1997, 1996, 1987, 1998, 2003],
    "gender": ["female", "male", "male", "male", "male", "female"],
    "city": ["New York", "Chicago", "Austin", "New York", np.nan, "New York"]
})

df1
```

```python
# No 2 Membuat Dataframe lainnya untuk dimerger dengan dataframe sebelumnya
df2 = pd.DataFrame({
    "nick_name": ["Anna", "David", "Dane", "Rose"],
    "full_name": ["Anna Levis", "David John", "Dane Sebastian", "Rose Kimberly"]
})

# Merger 2 dataframe dengan tetap mempertahankan baris pada dataframe originalnya
df = df1.merge(df2, how = "outer", left_on = "name", right_on = "nick_name")

df
```

```python
# No 3 Menghapus kolom nick_name dan mengisi nilai kosong di kolom full_name dengan kolom name yang asli
df = df.drop(columns = "nick_name")
df["full_name"] = df["full_name"].fillna(df["name"])

df
```

```python
# No 4 Menghitung bmi dari setiap orang dan membuat kolom baru bernama bmi
df["bmi"] = df["weight"] / (df["height"]/100)**2

df
```

```python
# No 5 mengklasifikasikan kelas dari bmi dan membuat kolom baru bernama bmi_class
def bmi_classification(bmi):
    if bmi >= 40:
        return 'Obesity class 3'
    elif bmi >= 35:
        return 'Obesity class 2'
    elif bmi >= 30:
        return 'Obesity class 1'
    elif bmi >= 25:
        return 'Overweight'
    elif bmi >= 18.5:
        return 'Normal weight'
    elif bmi < 18.5:
        return 'Underweight'
    else:
        return np.nan

# Menerapkan fungsi bmi_classification ke dataframe
df["bmi_class"] = df.apply(lambda x: bmi_classification(x.bmi), axis=1)

df
```
